```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, SimpleDirectoryReader, VectorStoreIndex
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor
from transformers import AutoModelForCausalLM, AutoTokenizer
import accelerate
```

```python
Settings.embed_model = HuggingFaceEmbedding(model_name = "BAAI/llm-embedder", token ="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")

Settings.llm = None
Settings.chunk_size = 256
Settings.chunk_overlap = 25
```

```python
documents = SimpleDirectoryReader("/kaggle/input/footballrules/articles").load_data()
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
top_k = 5
retriever = VectorIndexRetriever(index=index, similarity_top_k=top_k)
```

```python
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.5)]
)
```

```python
query = 'What is the offside rule?'
response = query_engine.query(query)
```

```python
print(response)
```

```python
model_name = "mistralai/Mistral-7B-Instruct-v0.2"
model = AutoModelForCausalLM.from_pretrained(model_name, token = "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", device_map = 'auto')
tokenizer = AutoTokenizer.from_pretrained(model_name, token = "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")
```

```python
prompt_template_w_context = lambda context,comment: f"""RulesGPT is an AI assistant that helps you understand the rules of soccer.
You can ask it questions about the rules of soccer, and it will provide you with the answers.
It gives you the information in a concise and easy to understand way.
Please respond to the following comment.
{context}
Please respond to the following comment. Use the above context if it i helpful
{comment}"""
```

```python
comment = "What is the offside rule?"
context = query_engine.query(comment)

prompt  = prompt_template_w_context(comment, context)
inputs = tokenizer(prompt, return_tensors="pt")
outputs = model.generate(input_ids=inputs["input_ids"], max_new_tokens=280)
print(tokenizer.batch_decode(outputs)[0])
```
